# Preparando Conexão com o Twitter

```python
# Imports Tweepy - biblioteca de python pra acesso à API do Twitter
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
```

```python
# Demais imports a serem utilizados nessa seção
from datetime import datetime
import json
```

```python
# Keys de acesso a minha API do Twitter
consumer_key = "XXXXXXX"
consumer_secret_key = "XXXXXXX"
access_token = "XXXXXXX-XXXXXXX"
access_token_secret = "XXXXXXX"
```

```python
# Acessando a API
auth = OAuthHandler(consumer_key, consumer_secret_key)
auth.set_access_token(access_token, access_token_secret)
```

```python
# Criando uma classe para capturar o stream dos dados e armazenar no MongoDB
class MyStreamListener(StreamListener):
    def on_data(self, data):
        tweet = json.loads(data)
        created_at = tweet["created_at"]
        id_str = tweet["id_str"]
        text = tweet["text"]
        obj = {"created_at":created_at, "id_str": id_str, "text":text,}
        tweetind = ead.insert_one(obj).inserted_id
        print(obj)
        return True
```

```python
# Criando um objeto da classe
myStreamListener = MyStreamListener()
```

```python
# Criando o objeto que possui a conexão com a API do twitter, juntamente com o acesso ao banco
myStream = Stream(auth, listener = myStreamListener)
```

# Preparando Conexão com o MongoDB

```python
# Import do módulo MongoClient para acessar o MongoDB dentro desse notebook
from pymongo import MongoClient
```

```python
# Estabelecendo conexão local com o MongoDB
client = MongoClient('localhost', 27017)
```

```python
# Acessando o banco de dados
db = client.tweetsdb
```

```python
# Criando a collection "ead"
ead = db.tweets
```

# Stream de Tweets

```python
# Palavras chaves que iremos buscar nos Tweets
keywords = ['ead', 'faculdade', 'universidade', 'aula']
```

```python
# Adicionando o filtro e gravando os tweets no banco
myStream.filter(track=keywords)
```

```python
# Fechando a conexão do objeto Stream
myStream.disconnect()
```

# Analisando os Tweets

```python
# Criando um dataset com os dados retornados do MongoDB
dataset = [{"created_at": item["created_at"], "text": item["text"],} for item in ead.find()]
```

```python
# Import dos módulos a serem utilizados nessa seção
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import sklearn
```

```python
# Utilizando o dataset para gerar um dataframe
df = pd.DataFrame(dataset)
```

```python
# Observando como está nosso dataframe
df
```

```python
# Gerando uma matriz de documentos
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# Separando os tweets por palavras
word_count = pd.DataFrame(cv.get_feature_names(), columns=["word"])
```

```python
# Contando o número de ocorrência das palavras
word_count["count"] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values("count", ascending=False).reset_index(drop=True)
```
